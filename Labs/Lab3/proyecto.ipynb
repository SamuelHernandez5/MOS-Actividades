## 1. Formulación Matematica

### Conjuntos

Centros de distribución:
$$D = \{D_1, D_2, D_3\}$$

Clientes: 
$$C = \{C_1, C_2, C_3\}$$

Nodos:
$$N = \{N_1, N_2, N_3, N_4, N_5, N_6\}$$
$$D ⊂ N$$
$$C ⊂ N$$

Vehiculos: 
$$K = \{K_1, K_2, K_3\}$$

### Parámetros

- Capacidad de almacenamiento de centro de distribución: $$a_i, \quad \forall i \in D$$
- Demanda de cliente : $$d_j, \quad \forall j \in C$$
- Capacidad de vehiculo: $$c_k, \quad \forall k \in K$$
- Rango de vehiculo: $$r_k, \quad \forall k \in K$$
- Distancia entre dos nodos: $$dis_{i,j}, \quad \forall i \in N \quad \forall j \in N$$
- Precio de rodaje por kilometro: $$P$$

### Variables de decisión

Asignación de inventario: Producto que se envia del centro i al cliente j
$$ z_{ij} \in {0,1}$$

Carga entregada en j por el vehiculo k
$$q_{jk} \geq 0$$

Ruta de los vehiculos:
$$x_{ijk} \in 0,1$$

Auxiliar para subtours
$$ u_{i} \leq n-1$$

### Restricciones

Cada nodo tiene una entrada
$$ \sum_{i\in D, i \neq j} \sum_{k\in K} x_{ijk} = 1  , \forall j \in 
N$$

Cada nodo tiene una salida
$$ \sum_{j\in O, i \neq j} \sum_{k\in K} x_{ijk} = 1 , \forall i \in N$$

No tener autociclos
$$ x_{i,i,k} = 0, \forall i \in N \forall k \in K $$

$$ u_{i} \geq 0$$

Subtours
$$ u_i - u_j + n \cdot x_{ijk} \leq n - 1, \forall i, j, k / i \neq j, 
i \neq 1, j \neq 1 \in N $$

Unicidad de localidad por equipo
$$ \sum_{K\in K} x_{ijk} = 1 , \forall j,i \in N / i \neq j$$

Capacidad de los centros de distribución:
$$\sum_{j\in C} z_{ij} \leq a_i , \forall i \in D$$

Demanda de los clientes:
$$\sum_{i\in D} z_{ij} = d_j , \forall j \in C$$

Capacidad del vehiculo
$$\sum_{j\in C} q_{jk} \leq c_k , \forall k \in K$$

Rango del vehiculo
$$suma de lo que recorre \leq r_k , \forall k \in K$$
